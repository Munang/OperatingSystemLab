```python
result=[]
count = len(txtfilenames)
cci = 0
for txtfilename in txtfilenames:
    
    binaryline=[]

    with open ("D:\\Drebin\\dex_exper\\dex_bin\\"+txtfilename+"","r") as f:
        lines = f.readlines()
        
        for each_line in lines:
            binaryline.append(each_line[10:63])
        
        ## 각각의 라인에서 이진코드만 추출한다 -> binaryline으로 대입

        bin_pixel=[]
        
        for binary in binaryline:
            bin_pixel.append((binary.split()))
  
        ## binaryline에 더해진 이진 코드를 8비트씩 떼어서 배열로 삽입
        
        bin_pixel2 = []

        
        ##bin_pixel2=sum(bin_pixel,[])
        for i in bin_pixel:
            bin_pixel2+=i
        ## 2차원 배열을 1차원 배열로 늘리기 
        
        pixel_array=[]
        
        for bin_pixel_s in bin_pixel2:
            str(bin_pixel_s)
            pixel_array.append(int(bin_pixel_s,2))
            
        ## 8비트의 이진 코드를 십진수로 변환 <255 의 숫자로 변환됨
        
        
#         result.append(pixel_array)
#         result.insert(0,txtfilename+'.dex')
        
        size =len(pixel_array)

        if (size<10240):
            width = 32

        elif (10240 <= size <= 10240*3):
            width = 64
        elif (10240*3 <= size <= 10240*6):
            width = 128
        elif (10240*6 <= size <= 10240*10):
            width = 256
        elif (10240*10 <= size <= 10240*20):
            width = 384
        elif (10240*20 <= size <= 10240*50):
            width = 512
        elif (10240*50 <= size <= 10240*100):
            width = 768
        else :
            width = 1024

        height = int(size/width)+1
        image = Image.new('L',(width,height))
        
        image.putdata(pixel_array)
        
        cci += 1
        image.save('D:\\Drebin\\dex_exper\\dex_img_rgb\\'+txtfilename[:-8]+'.jpg')
        print("\rcompleted : %d" % cci, ' 개 || 남은 갯수 : %d' % (count - cci), ' 개', end="")
        
```
